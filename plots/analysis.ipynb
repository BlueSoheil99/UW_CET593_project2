```python
df_normal = pd.read_csv('../Results/Normal Demand/UW_normal.csv')
df_decrease = pd.read_csv('../Results/Decrease Demand/UW_decrease.csv')
df_increase = pd.read_csv('../Results/Increase Demand/UW_increase.csv')
df_normal['Demand'] = 'Normal'
df_decrease['Demand'] = 'Low'
df_increase['Demand'] = 'High'
```
